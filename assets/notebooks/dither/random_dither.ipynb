```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
with open('KGF.dat') as f:
    image_original = np.loadtxt(f, dtype=int)
```

```python
plt.close()
plt.imshow(image_original, cmap='gray')
plt.axis('off')
plt.tight_layout()
```

```python
#thresholding
threshold = np.mean(image_original)
image_threshold = np.where(image_original > threshold, 1, 0)
plt.close()
plt.imshow(image_threshold, cmap='gray')
plt.axis('off')
plt.tight_layout()
#plt.savefig('KGF_threshold.png', dpi=300)
```

```python
#random dither
random_pixel = np.random.randint(0, 256, size=image_original.shape)
image_random_dither = np.where(random_pixel < image_original, 1, 0)
plt.close()
plt.imshow(image_random_dither, cmap='gray')
plt.axis('off')
plt.tight_layout()
#plt.savefig('KGF_random_dither.png', dpi=300)
```

```python
#this "error" is minimized for fixed thresholding
print(np.mean(np.abs(image_original/256 - image_threshold)))
print(np.mean(np.abs(image_original/256 - image_random_dither)))
```
