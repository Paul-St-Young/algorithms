```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
with open('KGF.dat') as f:
    image_original = np.loadtxt(f, dtype=int)
```

```python
#trim off a row so that shape is divisible by 4
image_original = np.copy(image_original[:-1,:])
x, y = image_original.shape
```

```python
plt.imshow(image_original, cmap='gray')
```

```python
#optimal pattern (?)
pattern = np.array([[1,9,3,11],[13,5,15,7],[4,12,2,10],[16,8,14,6]])
print(pattern)
```

```python
image_patterned = np.zeros_like(image_original)
for i in range(x//4):
    for j in range(y//4):
        image_patterned[4*i:4*i+4, 4*j:4*j+4] = image_original[4*i:4*i+4, 4*j:4*j+4]//16 > pattern
        #np.random.shuffle(pattern)
```

```python
plt.close()
plt.imshow(image_patterned, cmap='gray')
plt.axis('off')
plt.tight_layout()
#plt.savefig('KGF_ordered_dither.png', dpi=300)
```

```python
print(np.mean(np.abs(image_original/256 - image_patterned)))
```
